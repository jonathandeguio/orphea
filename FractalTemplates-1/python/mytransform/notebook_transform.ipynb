```python
from bosler import ReadDataFrame, WriteDataFrame
import pyspark.sql.functions as F
```

```python
source_dataset1 = "/Projects/Test/Data/Simple_Pipeline/SalesRecords"
target_dataset = "/Projects/Test/Data/Simple_Pipeline/SalesRecords_NB"
```

```python
df = ReadDataFrame(source_dataset1)
df.show()
```

```python
target_df = df.filter(df.Payment_Type == "Visa")
```

```python
target_df = target_df.withColumn("Transaction_date", F.to_timestamp(target_df["Transaction_date"], "M/d/yyyy HH:mm"))
```

```python
target_df.show()
```

```python
WriteDataFrame(target_df, target_dataset)
```
