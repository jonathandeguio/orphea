## Visualisation de données

Une partie essentielle de la science des données est la visualisation. Utilisons `matplotlib`, une bibliothèque de visualisation populaire, pour tracer un graphique simple.

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
